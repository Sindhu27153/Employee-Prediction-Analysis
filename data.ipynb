```python
# Encode target variable
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
df
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Drop unnecessary columns # multicolinearity columns
df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber','JobLevel','PercentSalaryHike','YearsAtCompany','MonthlyIncome','YearsInCurrentRole'], axis=1, inplace=True)
```

```python
df.columns
```

```python
# One-hot encode categorical features
df = pd.get_dummies(df, columns=['BusinessTravel', 'Department', 'EducationField', 
                                 'Gender', 'JobRole', 'MaritalStatus', 'OverTime'], 
                    drop_first=True)
```

```python
#Normalize The Numerical Columns
num_cols = ['Age', 'DailyRate', 'DistanceFromHome', 'HourlyRate', 
            'MonthlyRate', 'NumCompaniesWorked', 
            'TotalWorkingYears', 'TrainingTimesLastYear','YearsSinceLastPromotion', 'YearsWithCurrManager']

df[num_cols] = scaler.fit_transform(df[num_cols])

```

```python
# Define features (X) and target (y)
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
df['Attrition'].isna()
```

```python
# Split into train & test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#logistic regression # binary class
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score

```

```python
algorithms = {
    "LogisticRegression":LogisticRegression(max_iter=500),
    "lassoclassifier" :LogisticRegression(penalty='l1', solver='liblinear', C=1.0),
    "decisionTree" : DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42),
    "decisionTree" : DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=42),
    "RandomForest" : RandomForestClassifier(n_estimators=100),
    "ExtremeBoosting" : XGBClassifier(n_estimators=100, learning_rate=0.1),
    "Bagging" :BaggingClassifier(n_estimators=100, random_state=42),
    "NaiveBayes": GaussianNB(),
    "svc" :SVC()
}

results = []
```

```python
for name,model in algorithms.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    

    results.append({
        "Model": name,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    })
```

```python
#Model perfomance
print(pd.DataFrame(results))
```
